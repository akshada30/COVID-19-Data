```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://en.wikipedia.org/wiki/COVID-19_pandemic_in_India"
```

```python
response =requests.get(url)
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
table = soup.find('table', attrs={'class':'wikitable'})
```

```python
trs1 = table.select("tbody tr")[3:21]
trs2 = table.select("tbody tr")[22:39]
trs =trs1+trs2
```

```python
State = []
Cases = []
Deaths = []
Recoveries = []
Active = []
```

```python
import re
for tr in trs:
    th = tr.find("th", attrs={'scope': 'row'})
    State.append(th.find('a').text)
    
    tds = tr.find_all("td")
    c=tds[0].text.replace('\n', '').strip()
    if (re.search("[a-z]", c)):
        c=c[:-3]
    Cases.append(c.replace(",", "").strip())
    
    d=tds[1].text.replace('\n', '').strip()
    if (re.search("[a-z]", d)):
        d=d[:-3]
    Deaths.append(d.replace(",", "").strip())
    
    r=tds[0].text.replace('\n', '').strip()
    if (re.search("[a-z]", r)):
        r=r[:-3]
    Recoveries.append(r.replace(",", "").strip())
    
    a=tds[0].text.replace('\n', '').strip()
    if (re.search("[a-z]", a)):
        a=a[:-3]
    Active.append(a.replace(",", "").strip())
```

```python
data = list(zip(State, Cases, Deaths, Recoveries, Active))
```

```python
India_covid = pd.DataFrame(data, columns=['Name of State / UT', 'Cases', 'Deaths', 'Recoveries', 'Active'])
```

```python
India_covid.Cases = India_covid.Cases.astype(int)
India_covid.Deaths = India_covid.Deaths.astype(int)
India_covid.Recoveries = India_covid.Recoveries.astype(int)
India_covid.Active = India_covid.Active.astype(int)
```
